```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.utils import to_categorical
import time
import gc
from IPython.display import display
import ipywidgets as widgets
import warnings
warnings.filterwarnings('ignore')

# Dictionary of available models
available_models = {
    'MobileNetV2': MobileNetV2}

# Dictionary of available datasets and their properties
available_datasets = {
    'cifar10': {'num_classes': 10, 'image_shape': (32, 32, 3)}}

# Function to load and preprocess dataset
def load_and_preprocess_dataset(dataset_name, batch_size, img_size=96):
    print(f"Loading {dataset_name} dataset...")

    # Get dataset info
    num_classes = available_datasets[dataset_name]['num_classes']

    # Load dataset using TensorFlow Datasets
    if dataset_name == 'cifar10':

        (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

        # Convert labels to categorical
        y_train = to_categorical(y_train, num_classes)
        y_test = to_categorical(y_test, num_classes)

        # Get class names
        class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
                       'dog', 'frog', 'horse', 'ship', 'truck']

         # Calculate mean and std for standardization
        mean = np.mean(x_train, axis=(0, 1, 2))
        std = np.std(x_train, axis=(0, 1, 2))
        std = np.maximum(std, 1e-7)  # Prevent division by zero

        print(f"Mean pixel values: {mean}")
        print(f"Std pixel values: {std}")


        # Create preprocessing function
        def preprocess_image(image, label):
            # Cast to float32
            image = tf.cast(image, tf.float32)
            # Standardize
            image = (image - mean) / std
            # Resize
            image = tf.image.resize(image, (img_size, img_size))
            return image, label

        # Create tf.data datasets
        train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
        test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))

        # Apply preprocessing and create efficient batched datasets
        train_dataset = train_dataset.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
        train_dataset = train_dataset.cache().shuffle(10000).batch(batch_size).prefetch(tf.data.AUTOTUNE)

        test_dataset = test_dataset.map(preprocess_image, num_parallel_calls=tf.data.AUTOTUNE)
        test_dataset = test_dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

        # Create validation dataset for prediction visualization (25 images)
        validation_images = x_test[:25]
        validation_images = (validation_images - mean) / std
        validation_images = np.array([tf.image.resize(img, (img_size, img_size)).numpy() for img in validation_images])
        validation_labels = y_test[:25]

        return train_dataset, test_dataset, class_names, x_test.shape[0], validation_images, validation_labels


# Function to visualize model architecture using visualkeras
def visualize_model_architecture(model, model_name, dataset_name, output_path=None):
    print(f"Visualizing architecture for {model_name} on {dataset_name}...")

    try:
        # Check if visualkeras is installed, if not install it
        try:
            import visualkeras
        except ImportError:
            print("Installing visualkeras package...")
            import IPython
            IPython.get_ipython().system('pip install -q visualkeras pydot pillow')
            import visualkeras

        from PIL import ImageFont

        # Determine color mapping for layers
        color_mapping = {
            'Conv2D': '#2ECC71',            # Green
            'MaxPooling2D': '#3498DB',      # Blue
            'AveragePooling2D': '#9B59B6',  # Purple
            'BatchNormalization': '#F1C40F', # Yellow
            'Dropout': '#E74C3C',           # Red
            'Dense': '#1ABC9C',             # Teal
            'Flatten': '#34495E',           # Dark blue
            'Input': '#ECF0F1',             # Light gray
            'Add': '#95A5A6',               # Gray
            'GlobalAveragePooling2D': '#D35400', # Orange
            'Lambda': '#7F8C8D'             # Gray
        }

        # Use default color for unlisted layer types
        default_color = '#BDC3C7'  # Light gray

        # Try to load a font
        try:
            font = ImageFont.truetype("arial.ttf", 12)
        except:
            try:
                font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf", 12)
            except:
                font = None

        # Create the visualization - layered view
        architecture_path = f"{model_name}_{dataset_name}_architecture.png" if output_path is None else output_path
        architecture_img = visualkeras.layered_view(
            model,
            legend=True,
            color_map=color_mapping,
            to_file=architecture_path,
            draw_volume=True,
            spacing=20,
            font=font)

        # Display the architecture
        plt.figure(figsize=(10, 15))
        plt.imshow(architecture_img)
        plt.axis('off')
        plt.title(f"{model_name} Architecture for {dataset_name}")
        plt.tight_layout()
        plt.show()

        print(f"Architecture visualization saved to {architecture_path}")

    except Exception as e:
        print(f"Error during architecture visualization: {e}")
        import traceback
        traceback.print_exc()

# Build a model with pre-trained network
def build_model(model_name, input_shape, num_classes):
    print(f"Building {model_name} model for {num_classes} classes...")

    # Select base model
    base_model = available_models[model_name](
        weights='imagenet',
        include_top=False,
        input_shape=input_shape
    )

    # Freeze base model
    base_model.trainable = False

    # Define inputs
    inputs = tf.keras.Input(shape=input_shape)
    x = base_model(inputs)

    # Add classification head
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)

    predictions = Dense(num_classes, activation='softmax')(x)

    # Final model
    model = Model(inputs=inputs, outputs=predictions)

    # Compile with a lower learning rate to prevent NaN issues
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    model.summary()
    return model

# Train the model with gradient clipping to prevent NaN
def train_model(model, train_dataset, test_dataset, steps_per_epoch, validation_steps, epochs=15):
    print("\nTraining the model...")
    start_time = time.time()

    # Create a learning rate scheduler
    lr_schedule = tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=2,
        min_lr=1e-6,
        verbose=1
    )

    # Add early stopping
    early_stopping = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    )

    # Train with gradient clipping to prevent NaN issues
    model.optimizer.clipnorm = 1.0  # Clip gradients to prevent explosion

    # Checkpoint callback to save the best model
    checkpoint = tf.keras.callbacks.ModelCheckpoint(
        'best_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )

    # Train the model
    history = model.fit(
        train_dataset,
        epochs=epochs,
        validation_data=test_dataset,
        steps_per_epoch=steps_per_epoch,
        validation_steps=validation_steps,
        callbacks=[lr_schedule, early_stopping, checkpoint],
        verbose=1
    )

    training_time = time.time() - start_time
    print(f"\nTraining completed in {training_time/60:.2f} minutes.")

    return history, model

# Plot training history - accuracy
def plot_training_history(history, model_name, dataset_name):
    plt.figure(figsize=(8, 5))

    # Check if history has accuracy values
    if len(history.history.get('accuracy', [])) > 0:
        # Plot only accuracy
        plt.plot(history.history['accuracy'], label='Training')
        plt.plot(history.history['val_accuracy'], label='Validation')
        plt.title(f'{model_name} on {dataset_name} - Accuracy')
        plt.ylabel('Accuracy')
        plt.xlabel('Epoch')
        plt.legend()

        plt.tight_layout()
        plt.savefig(f'{model_name}_{dataset_name}_training_history.png')
        plt.show()
    else:
        print("No training history to plot or history contains NaN values.")

# Evaluate model and check predictions
def evaluate_model(model, test_dataset, class_names, validation_images, validation_labels, model_name, dataset_name):
    print(f"\nEvaluating {model_name} on {dataset_name} test data...")

    # Evaluate on test dataset
    test_results = model.evaluate(test_dataset, verbose=1)
    print(f"Test loss: {test_results[0]:.4f}")
    print(f"Test accuracy: {test_results[1]:.4f}")

    # Check for NaN in evaluation results
    if np.isnan(test_results[1]):
        print("WARNING: Test accuracy is NaN. Checking model predictions...")
        return float('nan')

    # Predict on validation images
    predictions = model.predict(validation_images)

    # Get predicted and true classes
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(validation_labels, axis=1)

    # Check if predictions contain NaN
    if np.any(np.isnan(predictions)):
        print("WARNING: Predictions contain NaN values!")

    # 5x5 grid
    fig, axes = plt.subplots(5, 5, figsize=(15, 15))
    fig.suptitle(f"Predictions on {dataset_name} with {model_name}", fontsize=16)

    for i, ax in enumerate(axes.flat):
        if i < len(validation_images):
            # Denormalize image for display
            img_to_show = validation_images[i] * 0.5 + 0.5  # Convert from [-1, 1] to [0, 1]
            img_to_show = np.clip(img_to_show, 0, 1)

            # For grayscale images, use cmap
            if img_to_show.shape[-1] == 1:
                ax.imshow(np.squeeze(img_to_show), cmap='gray')
            else:
                ax.imshow(img_to_show)

            pred_class = predicted_classes[i] if i < len(predicted_classes) else -1
            true_class = true_classes[i]

            # Show class names if available, otherwise show class numbers
            true_class_name = class_names[true_class] if true_class < len(class_names) else f"Class {true_class}"
            pred_class_name = class_names[pred_class] if pred_class < len(class_names) else f"Class {pred_class}"

            # Set title color based on correctness
            color = "green" if pred_class == true_class else "red"

            ax.set_title(f"True: {true_class_name}\nPred: {pred_class_name}", color=color, fontsize=8)
            ax.axis('off')

    plt.tight_layout()
    plt.subplots_adjust(top=0.9)  # Adjust for the suptitle
    plt.savefig(f'{model_name}_{dataset_name}_predictions.png')
    plt.show()

    # Calculate and display confusion matrix stats
    correct_predictions = sum(pred == true for pred, true in zip(predicted_classes, true_classes))
    accuracy = correct_predictions / len(true_classes)
    print(f"Prediction accuracy on sample: {accuracy:.4f} ({correct_predictions}/{len(true_classes)})")

    return test_results[1]

# Function to create UI for dataset and model selection
def get_user_selections():
    # Create dropdowns for model and dataset selection
    model_dropdown = widgets.Dropdown(
        options=list(available_models.keys()),
        value='MobileNetV2',
        description='Select Model:',
        style={'description_width': 'initial'},
        layout={'width': 'auto'}
    )

    dataset_dropdown = widgets.Dropdown(
        options=list(available_datasets.keys()),
        value='cifar10',
        description='Select Dataset:',
        style={'description_width': 'initial'},
        layout={'width': 'auto'}
    )

    # Input for batch size and epochs
    batch_size_input = widgets.IntSlider(
        value=64,
        min=16,
        max=256,
        step=16,
        description='Batch Size:',
        style={'description_width': 'initial'},
        layout={'width': 'auto'}
    )

    epochs_input = widgets.IntSlider(
        value=15,
        min=5,
        max=50,
        step=5,
        description='Epochs:',
        style={'description_width': 'initial'},
        layout={'width': 'auto'}
    )

    # Create a section for dataset info
    dataset_info = widgets.HTML("Select a dataset to see details")

    # Function to update dataset info when selection changes
    def update_dataset_info(change):
        dataset_name = change['new']
        dataset_info_dict = available_datasets[dataset_name]
        info_html = f"""
        <h4>Dataset: {dataset_name}</h4>
        <p>Number of classes: {dataset_info_dict['num_classes']}</p>
        <p>Original image shape: {dataset_info_dict['image_shape']}</p>
        """
        dataset_info.value = info_html

    # Connect the update function to the dropdown
    dataset_dropdown.observe(update_dataset_info, names='value')

    # Display the widgets
    print("Please select a model and dataset for training:")

    selection_box = widgets.VBox([
        widgets.HBox([model_dropdown, dataset_dropdown]),
        widgets.HBox([batch_size_input, epochs_input]),
        dataset_info
    ])

    display(selection_box)

    # Create a button to start training
    train_button = widgets.Button(
        description='Start Training',
        button_style='success',
        tooltip='Click to start training with the selected options'
    )

    display(train_button)

    # Initialize a variable to store training status
    training_status = {'started': False}

    # Function to handle button click
    def on_button_clicked(b):
        if training_status['started']:
            print("Training already in progress")
            return

        training_status['started'] = True
        selected_model = model_dropdown.value
        selected_dataset = dataset_dropdown.value
        batch_size = batch_size_input.value
        epochs = epochs_input.value

        print(f"\nStarting training with {selected_model} on {selected_dataset}...")
        print(f"Batch size: {batch_size}, Epochs: {epochs}")

        # Run the training process
        train_selected_model(selected_model, selected_dataset, batch_size, epochs)

        training_status['started'] = False

    # Connect the button to the function
    train_button.on_click(on_button_clicked)

    # Update the dataset info for the initial selection
    update_dataset_info({'new': dataset_dropdown.value})

# Main function - To use model and dataset selection
def train_selected_model(model_name, dataset_name, batch_size, epochs):
    print(f"Starting training with {model_name} on {dataset_name}...")

    # Set random seeds for reproducibility
    np.random.seed(42)
    tf.random.set_seed(42)

    try:
        # Get image size based on model requirements
        img_size = 96   # Smaller size for faster training on MobileNetV2
        print(f"Using image size: {img_size}x{img_size}")

        # Load dataset
        train_dataset, test_dataset, class_names, num_test_samples, validation_images, validation_labels = load_and_preprocess_dataset(
            dataset_name, batch_size, img_size
        )

        # Print basic dataset info
        print(f"Dataset: {dataset_name}")
        print(f"Number of classes: {len(class_names)}")
        print(f"Class names: {class_names}")


        # Calculate steps per epoch
        if dataset_name == 'cifar10':
          steps_per_epoch = 50000  // batch_size
        validation_steps = max(1, num_test_samples // batch_size)

        # Get number of classes and input shape
        num_classes = available_datasets[dataset_name]['num_classes']
        original_shape = available_datasets[dataset_name]['image_shape']

        # Keep the original number of channels but resize to target size
        input_shape = (img_size, img_size, original_shape[2])

        # Build model with appropriate input shape and number of classes
        model = build_model(
            model_name,
            input_shape,
            num_classes
        )

        # Visualize model architecture before training
        visualize_model_architecture(
            model,
            model_name,
            dataset_name,
            f'{model_name}_{dataset_name}_architecture.png'
        )

        # Train model
        history, model = train_model(
            model,
            train_dataset,
            test_dataset,
            steps_per_epoch = steps_per_epoch,
            validation_steps=validation_steps,
            epochs=epochs
        )

        # Plot training history
        plot_training_history(history, model_name, dataset_name)

        # Clear memory
        gc.collect()

        # Evaluate model and show predictions
        accuracy = evaluate_model(
            model,
            test_dataset,
            class_names,
            validation_images,
            validation_labels,
            model_name,
            dataset_name
        )

        # Save model
        model_filename = f'{model_name}_{dataset_name}_model.h5'
        model.save(model_filename)
        print(f"Model saved as {model_filename} with accuracy: {accuracy:.4f}")

        # Clear model from memory
        del model
        gc.collect()
        tf.keras.backend.clear_session()

        print(f"\nTraining completed for {model_name} on {dataset_name}!")

    except Exception as e:
        print(f"Error during training: {e}")
        import traceback
        traceback.print_exc()


def main():
    # Check TensorFlow version
    print(f"TensorFlow version: {tf.__version__}")

    # Make sure appropriate packages are installed
    print("Installing required packages if needed...")
    try:
        import IPython
        IPython.get_ipython().system('pip install -q tensorflow-datasets visualkeras pydot pillow')
    except:
        print("Not running in IPython environment, please make sure required packages are installed")

    # Show selection UI
    get_user_selections()

if __name__ == "__main__":
    main()
```
